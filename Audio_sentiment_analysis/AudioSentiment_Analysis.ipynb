```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
from nltk.sentiment import SentimentIntensityAnalyzer
from textblob import TextBlob
from wordcloud import WordCloud
import matplotlib.pyplot as plt

```

```python

import nltk
nltk.download('punkt')
```

```python
# Example text
text = "The stale smell of old beer lingers. It takes heat to bring out the odor a cold dip restores health and zest a salt pickle. Taste fine with ham tacos al Pastore are my favorite a zestful food is the hot cross bun."

# Tokenize into words
words = word_tokenize(text)

# Tokenize into sentences
sentences = sent_tokenize(text)
```

```python

import nltk
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.lower() not in stop_words]
```

```python
fdist = FreqDist(filtered_words)
fdist.plot(30, cumulative=False)
plt.show()
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
sia = SentimentIntensityAnalyzer()
sentiment_score = sia.polarity_scores(text)

blob = TextBlob(text)
sentiment_textblob = blob.sentiment
```

```python
wordcloud = WordCloud().generate(' '.join(filtered_words))
plt.figure(figsize=(8, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
